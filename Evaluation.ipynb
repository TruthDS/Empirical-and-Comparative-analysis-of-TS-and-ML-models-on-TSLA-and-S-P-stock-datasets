```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mse_scores = {
    'TSLA':[9534.7708, 87506, 5454.81, 53.83],
    'S&P': [9247.31, 4.87759, 467484.54, 750.57]
}

r2_scores = {
    "TSLA": [-0.69324, -1.51, -0.00, 0.99],
    "S&P": [0.147, -0.6541, -0.03, 1.00]
}

models = ["ARIMA", "VAR", "SVR", "Random Forest"]

fig, (ax1, ax2)  = plt.subplots(1, 2, figsize = (12, 6))

x = np.arange(len(models))
width = 0.35
ax1.bar(x - width/2, mse_scores["TSLA"], width, label = "Tesla")
ax1.bar(x + width/2, mse_scores["S&P"], width, label = "S&P 500")
ax1.set_title("MSE Scores")
ax1.set_xlabel("Models")
ax1.set_ylabel("MSE (log scale)")
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.set_yscale("log")
ax1.legend()

ax2.bar(x - width/2, r2_scores["TSLA"], width, label = "Tesla")
ax2.bar(x + width/2, r2_scores["S&P"], width, label = "S&P 500")
ax2.set_title("R-Squared Scores")
ax2.set_xlabel("Models")
ax2.set_ylabel("R-Squared score")
ax2.set_xticks(x)
ax2.set_xticklabels(models)
ax2.set_ylim(-1, 1)
ax2.legend()

plt.tight_layout()
plt.show()
```
